##Aplikasi sederhana menghitung volume bola

```python
jari_jari = int(input("Masukkan jari-jari dari bola (cm): "))
pi = 22/7
volume_bola = 4/3 * pi * jari_jari**3
print(f"Volume bola adalah {volume_bola} cm^3.")
```

##Aplikasi sederhana menghitung volume kerucut

```python
jari_jari = int(input("Masukkan jari-jari dari kerucut (cm): "))
tinggi = int(input("Masukkan tinggi dari kerucut (cm): "))
pi =22/7
volume_kerucut = 1/3 * pi * jari_jari **2*tinggi
print(f"Volume kerucut adalah {volume_kerucut}cm^2")
```
